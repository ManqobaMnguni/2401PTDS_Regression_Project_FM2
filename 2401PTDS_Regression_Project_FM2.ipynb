## 5.2. Graphical EDA

The Graphical Exploratory Data Analysis (EDA) sub-section involves visualizing data to uncover patterns, trends, and relationships that are not immediately apparent from statistical summaries alone. By employing various plotting techniques, such as histograms, scatter plots, box plots, and heatmaps, we can visually explore the distribution of individual variables, compare different groups, and identify correlations and anomalies. This visual approach enhances our understanding of the dataset and helps in detecting underlying structures and trends that may inform further analysis.

#### 5.2.1. Histogram

```python
features = ['Average Temperature °C', 'Forestland', 'Urban population', 'Rural population', 'Savanna fires', 'Forest fires', 'Drained organic soils (CO2)', 'total_emission']
df[features].hist(figsize=(10,10));
```

#### 5.2.1. Box Plot

```python
areas_list = df['Area'].unique()
print(areas_list)
y_var_box = input("Enter the Area for the box plot: ")

filtered_data_box = df[df['Area'] == y_var_box]


filtered_data_box.plot(
    kind='box', 
    subplots=True, 
    sharey=False, 
    figsize=(35, 16)
)

plt.subplots_adjust(wspace=15.0) 
plt.show()
```

#### 5.2.2. Violin Plot

```python
print("Available x variables: Average Temperature °C, Forestland, Urban population, Rural population, Savanna fires, Forest fires  Drained organic soils (CO2), total_emission")
x_variable = input("Enter the x variable for the violin plot: ")
areas_list = df['Area'].unique()
print(areas_list)
y_var = input("Enter the Area for the violin plot: ")


# Validate user input
if x_variable not in ['Average Temperature °C', 'Forestland', 'Urban population', 'Rural population', 'Savanna fires', 'Forest fires', 'Drained organic soils (CO2)', 'total_emission']:
    print("Invalid input. Please enter a correct variable name.")
else:
    # Filter Data base on Area input
    filtered_data = df[df['Area'] == y_var]
    # Create a violin plot
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=x_variable, data=filtered_data)
    plt.title(f'Violin Plot of {x_variable} for {y_var}')
    plt.xlabel(x_variable)
    plt.ylabel(y_var)
    plt.show()
```

#### 5.2.3. Bar Graph

##### 5.2.3.1. Bar Graph per year

```python
print("Available x variables: Average Temperature °C, Forestland, Urban population, Rural population, Savanna fires, Forest fires  Drained organic soils (CO2), total_emission")
x_variable_bar = input("Enter the x variable for the Bar graph: ")
areas_list_bar = df['Area'].unique()
print(areas_list_bar)
y_var_bar = input("Enter the Area for the Bar graph: ")


# Validate user input
if x_variable not in ['Average Temperature °C', 'Forestland', 'Urban population', 'Rural population', 'Savanna fires', 'Forest fires', 'Drained organic soils (CO2)', 'total_emission']:
    print("Invalid input. Please enter a correct variable name.")
else:
    # Filter Data base on Area input
    filtered_data_bar = df[df['Area'] == y_var_bar]
    # Create a violin plot
    plt.figure(figsize=(16, 8))
    sns.barplot(x='Year', y=x_variable_bar, data=filtered_data_bar)
    plt.title(f'Bar Graph of {x_variable_bar} for {y_var_bar}')
    plt.xlabel('Year')
    plt.ylabel(x_variable_bar)
    plt.show()
```
