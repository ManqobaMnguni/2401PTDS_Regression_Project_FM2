## 5.2. Graphical EDA

The Graphical Exploratory Data Analysis (EDA) sub-section involves visualizing data to uncover patterns, trends, and relationships that are not immediately apparent from statistical summaries alone. By employing various plotting techniques, such as histograms, scatter plots, box plots, and heatmaps, we can visually explore the distribution of individual variables, compare different groups, and identify correlations and anomalies. This visual approach enhances our understanding of the dataset and helps in detecting underlying structures and trends that may inform further analysis.

#### 5.2.1. Histogram

```python
features = ['Savanna fires', 'Forest fires','Crop Residues','Rice Cultivation','Drained organic soils (CO2)',
                'Pesticides Manufacturing','Food Transport','Forestland','Net Forest conversion',
                'Food Household Consumption','Food Retail','On-farm Electricity Use','Food Packaging',
                'Fertilizers Manufacturing','IPPU','Manure applied to Soils','Manure left on Pasture',
                'Manure Management','Fires in organic soils','Fires in humid tropical forests','On-farm energy use',
                'Rural population','Urban population','Total Population - Male','Total Population - Female',
                'total_emission','Average Temperature °C']
df[features].hist(figsize=(20,20));
```

#### 5.2.1. Box Plot

```python
features = ['Savanna fires', 'Forest fires', 'Crop Residues', 'Rice Cultivation', 'Drained organic soils (CO2)',
            'Pesticides Manufacturing', 'Food Transport', 'Forestland', 'Net Forest conversion',
            'Food Household Consumption', 'Food Retail', 'On-farm Electricity Use', 'Food Packaging',
            'Fertilizers Manufacturing', 'IPPU', 'Manure applied to Soils', 'Manure left on Pasture',
            'Manure Management', 'Fires in organic soils', 'Fires in humid tropical forests', 'On-farm energy use',
            'Rural population', 'Urban population', 'Total Population - Male', 'Total Population - Female',
            'total_emission', 'Average Temperature °C']

# Prompt user to select a feature
print("Please choose a feature number from the following list, Example 1:")
for i, feature in enumerate(features, 1):
    print(f"{i}. {feature}")
    
# Input selection (you can validate the input in a more advanced way, here assuming user input is valid)
selection = int(input("\nEnter the number corresponding to the feature you want to plot: ")) - 1
selected_feature = features[selection]

# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x=selected_feature)

# Customize the plot
plt.title(f"Box plot of {selected_feature}")
plt.xlabel(selected_feature)
plt.ylabel("Value")
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

#### 5.2.2. Violin Plot

```python
features = ['Savanna fires', 'Forest fires', 'Crop Residues', 'Rice Cultivation', 'Drained organic soils (CO2)',
            'Pesticides Manufacturing', 'Food Transport', 'Forestland', 'Net Forest conversion',
            'Food Household Consumption', 'Food Retail', 'On-farm Electricity Use', 'Food Packaging',
            'Fertilizers Manufacturing', 'IPPU', 'Manure applied to Soils', 'Manure left on Pasture',
            'Manure Management', 'Fires in organic soils', 'Fires in humid tropical forests', 'On-farm energy use',
            'Rural population', 'Urban population', 'Total Population - Male', 'Total Population - Female',
            'total_emission', 'Average Temperature °C']

# Prompt user to select a feature
print("Please choose a feature number from the following list, Example 1:")
for i, feature in enumerate(features, 1):
    print(f"{i}. {feature}")

selection = int(input("\nEnter the number corresponding to the feature you want to plot: ")) - 1
selected_feature = features[selection]

# Create the box plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x=selected_feature)

# Customize the plot
plt.title(f"Box plot of {selected_feature}")
plt.xlabel(selected_feature)
plt.ylabel("Value")

# Show the plot
plt.show()
```
